```python
import pandas as pd
import requests
from io import StringIO

# Base URL for Iowa State University IEM ASOS data
base_url = "https://mesonet.agron.iastate.edu/cgi-bin/request/asos.py"

# Query parameters
params = {
    "station": "EGKK",        # ICAO for London Gatwick
    "tz": "UTC",
    "year1": 2024, "month1": 1, "day1": 1, "hour1": 0, "minute1": 0,
    "year2": 2024, "month2": 12, "day2": 31, "hour2": 23, "minute2": 59,
    "data": "metar",          # request raw METAR string
    "format": "onlycomma"     # CSV format
}

# Fetch data
response = requests.get(base_url, params=params)

# Check response
if response.status_code == 200:
    df = pd.read_csv(StringIO(response.text))
    print(df.head())
    df.to_csv("EGKK_METAR_2024.csv", index=False)
    print("✅ METAR data saved to EGKK_METAR_2024.csv")
else:
    print("❌ Request failed:", response.status_code)
```
